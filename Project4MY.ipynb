### Chapter 3 - Visualizing Errors
- Basic Errorbars
- Continuous Errors

```python
# Chapter 3
# Game wins std deviation
w_std = wins.std()
print("Std dev", w_std, "wins")
plt.errorbar(yr, braves.W, yerr = w_std, fmt='.k')
plt.axis([2000, 2024, 0, 200])
plt.title('Std Deviation of wins by year')
plt.xlabel('Year')
plt.ylabel('Wins')
plt.show()
```

```python
# Chapter 3
# Game wins continuous error
from sklearn.gaussian_process import GaussianProcessRegressor

model = lambda x: x * np.cos(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * np.sqrt(MSE)  # 2*sigma ~ 95% confidence region

plt.plot(yr, braves.W, 'x')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                 color='gray', alpha=0.5)
plt.axis([2000, 2024, 0, 200])
plt.title('Std Deviation of wins by year')
plt.xlabel('Year')
plt.ylabel('Wins')
plt.show()
```

### Chapter 4 - Density and Contour Plots
- Visualizing 3D Function

```python
# Chapter 4
# hit distribution
fig = plt.figure()
ax = plt.axes()

def f(x, y):
    return x*y

player = pd.read_csv("Shohei4.csv")

hit_dist = player.hit_distance_sc
hit_speed = player.launch_speed
hit_ang = player.launch_angle
X, Y = np.meshgrid(hit_ang, hit_speed)
Z = f(X, Y)
plt.contour(X, Y, Z, cmap='RdGy')
```

### Chapter 8 - Multiple Subplots
- By Hand
- Grid of subplots
- Complicated Arrangements

```python
# Chapter 8
# Multiple Subplots
import math

def height_at_time(t, ang, speed):
    g = 9.81
    ang_rad = math.radians(ang)
    height = (speed * np.sin(ang_rad) * t) - (0.5 * g * t**2)
    return height
def distance_at_time(t, ang, speed):
    ang_rad = math.radians(ang)
    distance = speed * np.cos(ang_rad) * t/2.5
    return distance

fig = plt.figure()
x = np.linspace(0, 10)
hr = player.loc[player.events == "home_run"]
print(hr.iloc[1])

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-5, 150))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-5, 500))

ax1.plot(height_at_time(x, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed))
ax2.plot(distance_at_time(x, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed))
plt.title('Random Home Run height and distance')
plt.xlabel('Time')
plt.ylabel('Distance/Height')
plt.show()
```

```python
# 10 Homerun height plots
fig = plt.figure()

for i in range(1,10):
    ax = fig.add_subplot(3, 3, i)
    ax.plot(x, height_at_time(x, hr.iloc[i].launch_angle, hr.iloc[i].launch_speed))
fig.text(x=0.3, y=0.9, s='First 10 Homeruns Height graph')
```

```python
#Complicated Arrangement
fig = plt.figure(figsize=(8,8))
grid = plt.GridSpec(3,3,hspace=0.2, wspace=0.2)
main = fig.add_subplot(grid[:-1,1:])
main.plot(hit_ang, hit_dist, 'ok', markersize=1, color='red')
xplt = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main)
yplt = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main)
xplt.hist(hit_dist, 30, orientation='horizontal')
yplt.hist(hit_ang, 30, orientation='vertical')
plt.show()
```

```python
# Fancy Ticks
def distance_at_time(t, ang, speed):
    ang_rad = np.radians(ang)
    distance = speed * np.cos(ang_rad) * t/2.5
    return distance

def format_feet(dist, tick_num):
    return f"{dist} ft"

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10)
hr = player.loc[player.events == "home_run"]

ax.plot(height_at_time(x, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed))
ax.yaxis.set_major_formatter(plt.FuncFormatter(format_feet))

plt.title('Random Home Run height and distance')
plt.xlabel('Time')
plt.show()
```

### Chapter 12 - Three-Dimensional Plotting in Matplotlib
- 3D Points and Lines
- 3D Contour Points
- Wireframes and Surface Plots
- Surface Triangulations

```python
# Chapter 12
# Three-Dimensional Plotting in Matplotlib
fig = plt.figure()
ax = plt.axes(projection='3d')

hr = player.loc[player.events == "home_run"]
angle = np.random.randint(90)
angle = np.radians(angle)
print(hr.iloc[1])

t = np.linspace(0, 10)
x = distance_at_time(t, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed) * np.sin(angle)
y = distance_at_time(t, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed) * np.cos(angle)
z = height_at_time(t, hr.iloc[1].launch_angle, hr.iloc[1].launch_speed)

ax.set_xlim(400, 0)
ax.set_ylim(0, 400)
ax.set_zlim(0, 200)
ax.plot3D(x, y, z)
plt.show()
```
