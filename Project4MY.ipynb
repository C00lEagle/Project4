### Chapter 1 - Simple Line Plots
- Line color and styles
- Axes limits
- Labeling plots
- Saving Figures

```python
#data from baseball reference.com
braves = pd.read_csv('braves.csv')
braves
```

```python
# Chapter 1
# Braves wins by the years
fig = plt.figure()
ax = plt.axes()

yr = braves.Year
wins = braves.W
losses = braves.L
games = braves.G
ax.plot(yr, wins.values, color='green', label='W')
ax.plot(yr, losses.values, color='red', label='L')
ax.plot(yr, games.values, label='GP', linestyle=':')
ax.legend()
fig.savefig('braveswins.png')

plt.axis([1870, 2000, 10, 170])
plt.title('Braves Wins From 1870 to 2000')
plt.xlabel('Year')
plt.ylabel('Games')
plt.show()
ax
```

### Chapter 2 - Simple Scatter Plots
- plt.plot vs plt.scatter

```python
# Chapter 2
# Braves Win% by Organization Name
diff_braves = pd.read_csv('diff_braves.csv')
diff_braves = diff_braves[:-1]
print(diff_braves)

plt.style.use('default')
plt.plot(diff_braves.index, diff_braves['W-L%'], 'x', color='blue')
plt.title(f'Braves Win% by Organization Name')
plt.xlabel('Dataframe Index')
plt.ylabel(f'Win%')
plt.show()
```
